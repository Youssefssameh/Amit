```python
class Calculator():
    def __init__(self):
        self.name = input("Please enter your name: ")
        print(f"Welcome {self.name} to Amit Calculator")

    def summation(self, *numbers):
        result = numbers[0]
        for num in numbers[1:]:
            result += num
        return result

    def subtraction(self, *numbers):
        result = numbers[0]
        for num in numbers[1:]:
            result -= num
        return result

    def multiplication(self, *numbers):
        result = numbers[0]
        for num in numbers[1:]:
            result *= num
        return result

    def division(self, *numbers):
        result = numbers[0]
        for num in numbers[1:]:
            if num == 0:
                return "Cannot divide by zero"
            else:
                result /= num
        return result

    def even_odd(self, num):
        if num % 2 == 0:
            return f"{num} is even"
        else:
            return f"{num} is odd"

    def factorial(self, num):
        if num == 0 or num == 1:
            return 1
        else:
            return num * self.factorial(num - 1)
        
    def run_calculator(self):
        while True:
            print("\n\nOperations: \n1. Addition\n2. Subtraction\n3. Multiplication\n4. Division\n5. Even or Odd\n6. Factorial\n7. Exit\n\n")
            choice = input("Enter your choice: ")
            if choice in ("1", "2", "3", "4"):
                numbers = [int(num) for num in input("Enter your numbers separated by comma: ").split(",")]
                if choice == "1":
                    print(f"Result: {self.summation(*numbers)}")
                elif choice == "2":
                    print(f"Result: {self.subtraction(*numbers)}")
                elif choice == "3":
                    print(f"Result: {self.multiplication(*numbers)}")
                elif choice == "4":
                    print(f"Result: {self.division(*numbers)}")
            elif choice in ("5", "6"):
                first_number = float(input("Enter your number: "))
                if choice == "5":
                    print(self.even_odd(first_number))
                elif choice == "6":
                    print(f"Factorial {first_number} = {self.factorial(first_number)}")
            elif choice == "7":
                print("Thank you, see you soon")
                break
            else:
                print("Invalid input, please try again")

#-------------------------------------------------------------------------------------------------
c1 = Calculator()
c1.run_calculator()
```
